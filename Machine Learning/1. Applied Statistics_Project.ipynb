# Problem Statement:
An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure  rate  for  the  manufactured  item  is  5%.  Suppose  a  random  sample  of  10  manufactured  items  is  selected.  Answer  the  following questions.

### 2.A. Probability that none of the items are defective?

```python
# As each trail has only two outcomes (defective / not defective) and fixed number of samples are given,
# we can assume Binomial Distributon here

import scipy.stats as stats

p = 0.05    # probability of a defective piece
n = 10      # Number of items in the sample
k = 0       # To find probability of number of items defective

p1 = stats.binom.pmf(k=k, n=n, p=p)

print("Probability that none of the items are defective is : %.4f " % p1)
```

### 2.B. Probability that exactly one of the items is defective?

```python
k = 1       # To find probability of 1 bulb being defective

p1 = stats.binom.pmf(k=k, n=n, p=p)

print("Probability that exactly one of the items is defective is : %.4f " % p1)
```

### 2.C. Probability that two or fewer of the items are defective?

```python
k = 2

# Here, we need to calculate cumulative probability
p1 = stats.binom.cdf(k=k, n=n, p=p)

print("Probability that two or fewer of the items are defective : %.4f " % p1)
```

### 2.D. Probability that three or more of the items are defective ?

```python
print("Probability that two or fewer of the items are defective : %.4f " % (1 - p1))
```

# Problem Statement:
A car salesman sells on an average 3 cars per week.

### 3.A. What is Probability that in a given week he will sell some cars?

```python
# In the given example, sucess rate is given but number of items in the sample is not fixed.
# So we can assume Poisson Distribution

rate = 3
k = 0       # Some cars means :  1 - (zero cars)

p = 1 - stats.poisson.cdf(k,rate)

print("Probability that in a given week he will sell some cars is : %.4f" % p)
```

### 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?

```python
p = stats.poisson.cdf(4,rate) - stats.poisson.cdf(1,rate)

print("Probability that in a given week he will sell 2 or more but less than 5 cars is : %.4f" % p)
```

### 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

```python
x = np.arange(0,11)
p = stats.poisson.cdf(x,rate)

plt.figure(figsize=(5,3))
plt.xlabel("Number of cars sold per week")
plt.ylabel("CDF")
plt.title("Cars sold per-week (CDF) vs number of cars sold per week")
sns.lineplot(x=x,y=p);
```

```python
k = 0

p1 = stats.binom.pmf(k=k, n=n, p=p)

print("Probability that none of the three orders will be recognised correctly is : %.4f" % p1)
```

## 4. Statistical Analysis

### A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?

```python
df.groupby('OperatingState')['Funds_in_million'].mean()
```

There is some difference between the mean of funds raised by the comapnies which are Operating and which are Closed,
but let's see if the differnce is significant.

We can check this using a 2-T independent sample test.

### B. Write the null hypothesis and alternative hypothesis.

Let A be the mean of funds raised by the comapnies which are Operating and B be the the mean of funds raised by the comapnies which are Closed.

H0 :  A  =  B
HA :  A      B

### C. Test for significance and conclusion

```python
from scipy.stats import ttest_ind

# df[df['OperatingState'] == 'Operating']
t_statistic, p_value  = ttest_ind(df[df['OperatingState'] == 'Operating']['Funds_in_million'],
                                  df[df['OperatingState'] == 'Closed']['Funds_in_million'])
print('tstat',t_statistic)
print('P Value',p_value)

alpha = 0.05

if p_value < alpha:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
    print('We conclude that there is significant difference between Funds raised by companies that are still operating vs companies that closed down.')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
    print('We conclude that there is no significant difference between Funds raised by companies that are still operating vs companies that closed down.')
```

### F. Calculate percentage of winners that are still operating and percentage of contestants that are still operating

```python
Winners = r['Winner']
contestants = r['Contestant']

r1 = df[df['OperatingState'] == 'Operating']['Result'].value_counts()
winners_operating = r1['Winner']
contestants_operating = r1['Contestant']

p_winners_operating = winners_operating / Winners
p_contestants_operating = contestants_operating / contestants

print(f"Percentage of winners that are still operating is : %.2f" % p_winners_operating)
print(f"Percentage of contestants that are still operating is : %.2f " % p_contestants_operating)
```

- There is some difference between percentage of winners that are still operating and percentage of contestants that are still operating, but let's see if the differnce is significant.
- We can use Z-test for proportions

### G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:


**Null hyputhesis (Ho):** The proportion of winners that are still operating and the propotion of contestants that are still operating is same.

**Alternative hypothesis (Ha)**: The proportion of winners that are still operating and the propotion of contestants that are still operating is not same.

```python
from statsmodels.stats.proportion import proportions_ztest

test_statistic, p_value = proportions_ztest([contestants_operating, winners_operating], [contestants, Winners])

print("P-value : ", p_value)

if p_value < alpha:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
    print('We conclude that the proportion of winners that are still operating and the propotion of contestants that are still operating is not same.')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
    print('We conclude that the proportion of winners that are still operating and the propotion of contestants that are still operating is same.')
```
